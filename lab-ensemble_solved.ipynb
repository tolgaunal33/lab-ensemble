```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor,AdaBoostRegressor, GradientBoostingRegressor, BaggingClassifier, RandomForestClassifier,AdaBoostClassifier, GradientBoostingClassifier

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

Now perform the same as before:
- Feature Scaling
- Feature Selection


```python
spaceship_cleaned = spaceship.dropna()
spaceship_cleaned['Cabin'] = spaceship_cleaned['Cabin'].str[0]
spaceship_cleaned.drop(columns=['PassengerId', 'Name'], inplace=True)
```

```python
categorical_cols = ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP']
spaceship_cleaned_encoded = pd.get_dummies(spaceship_cleaned, columns=categorical_cols, drop_first=True)
```

**Perform Train Test Split**

```python
#your code here

features = spaceship_cleaned_encoded.drop(columns = ["Transported"])
target = spaceship_cleaned_encoded["Transported"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
normalizer = MinMaxScaler()

normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python

X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

**Model Selection** - now you will try to apply different ensemble methods in order to get a better model

- Bagging and Pasting

```python
#your code here

bagging_reg = BaggingRegressor(DecisionTreeRegressor(max_depth=20),
                               n_estimators=100,
                               max_samples = 1000)
```

```python
bagging_reg.fit(X_train_norm, y_train)
```

```python
pred = BaggingClassifier(DecisionTreeClassifier(max_depth=20),
                        n_estimators=100,
                        max_samples = 1000)
```

```python
pred.fit(X_train_norm, y_train)
```

```python
pred.score(X_test_norm, y_test)
```

- Random Forests

```python
#your code here
forest_class = RandomForestClassifier(n_estimators=100,
                             max_depth=10, random_state=42)
```

```python
forest_class.fit(X_train_norm, y_train)
```

```python
forest_class.score(X_test_norm, y_test)
```

- Gradient Boosting

```python
#your code here

gb_reg = GradientBoostingClassifier(max_depth=20,
                                   n_estimators=100)
```

```python
gb_reg.fit(X_train_norm, y_train)
```

```python
gb_reg.score(X_test_norm, y_test)
```

- Adaptive Boosting

```python
#your code here

ada_reg = AdaBoostClassifier(DecisionTreeClassifier(max_depth=20),
                            n_estimators=100)
```

```python
ada_reg.fit(X_train_norm, y_train)
```

```python
ada_reg.score(X_test_norm, y_test)
```
